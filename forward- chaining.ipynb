```python
element = { 
        'A' : True,
        'B' : True,
        'C' : True,
        'D' : False,
        'E' : False,
        'F' : False,
        'G' : False,
        'H' : False,
        }
def printout():
  print('Now we have ')
  if element['A'] is True:
    print('A')
  if element['B'] is True:
    print('B')
  if element['C'] is True:
    print('C')
  if element['D'] is True:
    print('D')
  if element['E'] is True:
    print('E')
  if element['F'] is True:
    print('F')
  if element['G'] is True:
    print('G')
  if element['H'] is True:
    print('H')

new_fact = True

rules_fired = ['facts']

while new_fact == True: 
    
    # new cycle
    new_fact = False  
    
    
    # Rule 1 IF A & B THEN D 
    if element['A'] is True and element['B'] is True:
        element['D'] = True
        new_fact = True
        rules_fired.append('R1')
        print('\nRule 1:Because we have A and B, so we got D.')
        printout()
        

    # Rule 2 IF B & D THEN F 
    if element['B'] is True and element['D'] is True:
        element['F'] = True
        new_fact = True
        rules_fired.append('R2')
        print('\nRule 2:Because we have B and D, so we got F.')
        printout()

    # Rule 3 IF F THEN G 
    if element['F'] is True :
        element['G'] = True
        new_fact = True
        rules_fired.append('R3')
        print('\nRule 3:Because we have F, so we got G.\n')
        printout()

    # Rule 4 IF A & E THEN H 
    if element['A'] is True and element['E'] is True:
        element['H'] = True
        new_fact = True
        rules_fired.append('R4')
        print('\nRule 4:Because we have A and E, so we got H.\n')
        printout()
    
    # Rule 5 IF A & C THEN E 
    if element['A'] is True and element['C'] is True:
        element['E'] = True
        new_fact = True
        rules_fired.append('R5')
        print('\nRule 5:Because we have A and C, so we got E.\n')
        printout()

    # output classification result
    

    if element['H'] is True:
        print('We got element H !\n')    
        print('Rules fired: ')
        print(rules_fired)
        new_fact = False
```
